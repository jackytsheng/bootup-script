```python
from collections import Counter
from itertools import combinations

def find_frequent_combinations(data, combination_length,top_common=None):
    combinations_list = []

    for row in data:
        combinations_list.extend(list(combinations(row, combination_length)))

    count = Counter(combinations_list)
    most_common = count.most_common(top_common)

    return most_common
```

```python
from collections import defaultdict

def find_consecutive_strikes(data):
    cur_strike_map = defaultdict(int)
    strike_set_map=defaultdict(int)
    for i in range(1,len(data)):
        prev_row = data[i-1]
        current_row = data[i]

        for number in prev_row:
            if number in current_row:
                cur_strike_map[number] += 1
            elif number in cur_strike_map:
                strike_set_map[(number,cur_strike_map[number])] += 1
                cur_strike_map.pop(number)
        
    
    for k,v in cur_strike_map.items():
        strike_set_map[(k,v)]+=1
    
    return strike_set_map
```

```python
from collections import defaultdict

def find_not_consecutive_strikes(data):
    not_shown_strike_map = defaultdict(int)
    not_shown_strike_set_map = defaultdict(int)
    for row in data:
        for number in range(1,36):
            if number not in row:
                not_shown_strike_map[number] += 1
            elif not_shown_strike_map[number] != 0:
                not_shown_strike_set_map[(number,not_shown_strike_map[number])] += 1
                not_shown_strike_map.pop(number)


    
    
    return not_shown_strike_set_map, not_shown_strike_map
```

```python

import matplotlib.pyplot as plt
import math

def plot(data,barcolor="steelblue"):
    unique_titles = set(key[0] for key in data.keys())
    num_rows = 5
    num_cols = 7
    num_charts = num_rows * num_cols

    fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 10))

    chart_count = 0
    for title in unique_titles:
        filtered_data = {key[1]: value for key, value in data.items() if key[0] == title}

        x_values = list(filtered_data.keys())
        y_values = list(filtered_data.values())

        row = math.floor(chart_count / num_cols)
        col = chart_count % num_cols

        ax = axs[row, col]
        ax.bar(x_values, y_values,color=barcolor)
        ax.set_xlabel('Gap')
        ax.set_title(str(title))
        ax.set_yticks(range(max(y_values) + 1))
        ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

        chart_count += 1

    # Remove extra subplots if not used
    for i in range(chart_count, num_charts):
        row = math.floor(i / num_cols)
        col = i % num_cols
        fig.delaxes(axs[row, col])

    plt.tight_layout()
    plt.show()
```
